```python
# import xlsxwriter module
import xlsxwriter
import random
import copy

# Some data we want to write to the worksheet.
town_name = ["Town01", "Town02", "Town03", "Town04", "Town05", "Town06", "Town07", "Town08"]
weather_name = ["ClearNight", "ClearNoon", "ClearSunset", "FogMorning", "HardRainSunset", "WetCloudyNoon", "WetSunset"]
dir="conditions/groups/"

index_town_allweather = 6
cross_town_name = town_name[:index_town_allweather] + town_name[index_town_allweather+1:]

start_id = 0
participant_num = 50
for index in range(participant_num):
    filename_participant = "{id:02d}.xlsx".format(id=index+start_id)
    
    workbook = xlsxwriter.Workbook(dir+filename_participant)
    worksheet = workbook.add_worksheet()

    random_weather = random.shuffle(weather_name)
    
    # write town weather name
    worksheet.write(0, 0, "town_weather_name")
    
    front_town_name = town_name[:index_town_allweather]
    print(front_town_name)
    for i in range(0,len(front_town_name)) :
        item = town_name[i]+"_"+weather_name[i]
        print(item)
        worksheet.write(i+1, 0, item)
    
    weather_for_onetown_name = copy.deepcopy(weather_name)
    random_weather_for_onetown = random.shuffle(weather_for_onetown_name)
    # print(weather_for_onetown_name)
    for i in range(len(weather_for_onetown_name)) :
        item = town_name[index_town_allweather]+"_"+weather_for_onetown_name[i]
        print(item)
        worksheet.write(i+1+len(front_town_name), 0, item)

    behind_town_name = town_name[index_town_allweather+1:]
    print(behind_town_name)
    for i in range(len(behind_town_name)):
        item = behind_town_name[i]+"_"+weather_name[i+len(front_town_name)]
        print(item)
        worksheet.write(i+1+len(front_town_name)+len(weather_for_onetown_name), 0, item)

    workbook.close()
```
